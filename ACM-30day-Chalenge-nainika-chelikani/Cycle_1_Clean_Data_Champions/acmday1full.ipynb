```python
import pandas as pd
data= pd.read_csv('Desktop/acm s3/burnout dataset/mental_health_workplace_survey.csv')
print(data)
```

```python
#handling missing values
data.isnull().sum() 
##no missing values found so i got o in all columns hence dropping rows or columns is not necessary
```

```python
##removing outliers 
##in the dataset workhoursperweek is given so 18 hrs per day*7 i.e 126 hrs 
##so removing any row with workhoursperweek>126
data['WorkHoursPerWeek'].max()
```

```python
#the max value is 59 which is below the 126 so no need to remove any rows
#proving it visually
import numpy as np
import matplotlib.pyplot as plt
sorted = data['WorkHoursPerWeek'].sort_values().values
x_values = np.arange(len(sorted))
plt.figure(figsize=(10, 5))
plt.scatter(x_values, sorted, color='blue', label="Employee's Work Hours")
plt.axhline(y=126, color='red')
plt.title('Work Hours Per Week', fontsize=14)
plt.ylabel('Work Hours Per Week', fontsize=12)
plt.xlabel('Employee Index', fontsize=12)
plt.show()

```

```python
#from the graph the max is 59
data.columns
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['SleepHours'], data['StressLevel'], alpha=0.6, color='black')
plt.title('StressLevel vs SleepHours', fontsize=14)
plt.xlabel('SleepHours', fontsize=12)
plt.ylabel('StressLevel', fontsize=12)
plt.grid(True)
plt.show()
#this is a scatterplot
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.boxplot(x='HasMentalHealthSupport', y='BurnoutLevel', data=data)
plt.title('Burnout Level by Mental Health Support', fontsize=14)
plt.xlabel('Has Mental Health Support', fontsize=12)
plt.ylabel('Burnout Level', fontsize=12)
plt.show()
#this is a boxplot
```

```python
#The boxplot shows that burnout levels look pretty similar for people with or without mental health support. 
#It doesn’t seem like just having support automatically lowers burnout. There are probably other things going on that affect burnout too.

```
